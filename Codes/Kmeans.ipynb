```python
from pandas import read_csv
data = "/Users/mdarikrayhan/Documents/GitHub/CSE-4250-Data-Mining-and-Data-Ware-housing-Sessional/Codes/pima_indians_diabetes_data.csv"
#apply kmeans on the dataset
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = read_csv(data)
scaler = StandardScaler()
kmeans = KMeans(n_clusters=2)
pipeline = make_pipeline(scaler, kmeans)
pipeline.fit(df)
labels = pipeline.predict(df)
df['labels'] = labels
print(df.head())
#plot the clusters
plt.scatter(df['Glucose'], df['BloodPressure'], c=labels, alpha=0.5)
plt.xlabel('Glucose')
plt.ylabel('BloodPressure')
plt.show()
```
