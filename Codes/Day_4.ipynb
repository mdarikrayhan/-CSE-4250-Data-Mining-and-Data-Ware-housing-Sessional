### Code-3: Normalize Data with respect to age column

```python
print(data.head())

#separate age from the data
age=data['Age']
print(age)

max_age=age.max()
min_age=age.min()

#normalize age within 0 to 1
print('Max age: ',max_age)
print('Min age: ',min_age)
normalized_age=(age-min_age)/(max_age-min_age)
print(normalized_age)
```

### Code-4: Copy the normalized data to a new dataframe

```python
#new data frame with normalized age
new_data=data.copy()
new_data['Age']=normalized_age
print(new_data)
```

### Code-8:Correction Matrix Plot

```python
# Correction Matrix Plot
from matplotlib import pyplot
from pandas import read_csv
import numpy
data=pd.read_csv('/Users/mdarikrayhan/Documents/GitHub/CSE-4250-Data-Mining-and-Data-Ware-housing-Sessional/Codes/pima_indians_diabetes_data.csv')
correlations = data.corr()
# plot correlation matrix
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
pyplot.show()
```
