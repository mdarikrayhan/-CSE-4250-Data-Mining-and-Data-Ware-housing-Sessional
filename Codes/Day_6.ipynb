```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
from warnings import filterwarnings
filterwarnings(action='ignore')
```

```python
data = pd.read_csv("Lab_Work_Data.csv")
print(data)
```

```python
#checking for missing values
print(data.isnull().sum())
```

```python
#fill missing values in Income with mean
data['Income (JAN)'] = data['Income (JAN)'].fillna(data['Income (JAN)'].mean())
data['Income (FEB)'] = data['Income (FEB)'].fillna(data['Income (FEB)'].mean())
print(data)
```

```python
#Compare between Income (JAN) and Income (FEB) using histogram
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

axs[0].hist(data['Income (JAN)'])
axs[0].set_title('Income (JAN)')

axs[1].hist(data['Income (FEB)'])
axs[1].set_title('Income (FEB)')

plt.show()
```
